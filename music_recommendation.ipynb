```python
song_df=song_df.head(5000)
song_df.shape
```

```python
#combining artist and title name
song_df['song']=song_df['title'].astype(str)+ song_df['artist_name']
song_df.head()
```

```python
#cululamtive sum of listen count of the songs based on the songs 
song_grouped=song_df.groupby(['song']).agg({'listen_count':'count'}).reset_index()
song_grouped.head()
```

```python
grouped_sum=song_grouped['listen_count'].sum()
grouped_sum
```

```python
song_grouped['listen_count']
```

```python
song_grouped['percentage']=song_grouped['listen_count']/grouped_sum*100
song_grouped.sort_values(['listen_count','song'],ascending=[0,1])
```
